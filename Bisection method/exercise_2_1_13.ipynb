**Exercise 2.1.13** 

Find an approximation to $^3\sqrt{25}$ correct to within $10^{-4}$ using Bisection method.

```python
import numpy as np
import pandas as pd

#Initialization
a    = 2        #lower bound
b    = 3        #upper bound
TOL  = (10**-4)
nmax = 50

#function
def f(x):
  return (x**3)-25

#main
data = []
i    = 0
FA   = f(a)

while i<nmax:
  
  p  = a + (b-a)/2 #midpoint
  FP = f(p)

  out = [i+1,a,b,p,FP]
  data.append(out)  #output
  i   = i+1

  if FP == 0 or (b-a)/2<TOL:
    break #stop

  if FA*FP>0:
    a  = p
    FA = FP
  else:
    b=p

  
df = pd.DataFrame(data, columns=['Iterasi (n)', 'a_n','b_n','p_n','f(p_n)'])
display(df)
print(f'RESULT p_{i} = {p}')
```
