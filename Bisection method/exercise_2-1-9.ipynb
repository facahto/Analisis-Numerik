a. Sketch the graphs of $y=e^x-2$ and $y=\cos{(e^x-2)}$.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd

x  = np.arange(0,1.5,0.1)

y1 = np.exp(x)-2
y2 = np.cos(np.exp(x)-2)

plt.figure(figsize=(10, 7))
plt.plot(x, y1, color='g', label='exp(x)-2')
plt.plot(x, y2, color='b', label='cos(exp(x)-2)')
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(linestyle = '--', linewidth = 0.5)
plt.show()
```

b. Use the Bisection method to find an approximation to within $10^{-5}$ to a value in $[0.5,1.5]$ with $e^x-2=\cos{(e^x-1)}$.

Hint: $e^x-2-\cos{(e^x-1)}=0$

```python
#Initialization
a    = 0.5        #lower bound
b    = 1.5        #upper bound
TOL  = (10**-5)
nmax = 50

#function
def f(x):
  y1 = math.exp(x)-2
  y2 = math.cos(math.exp(x)-2)
  return y1 - y2 

#main
data = []
i    = 0
FA   = f(a)

while i<nmax:
  
  p  = a + (b-a)/2 #midpoint
  FP = f(p)

  out = [i+1,a,b,p,FP]
  data.append(out)  #output
  i   = i+1

  if FP == 0 or (b-a)/2<TOL:
    break #stop

  if FA*FP>0:
    a  = p
    FA = FP
  else:
    b=p

  
df = pd.DataFrame(data, columns=['Iterasi (n)', 'a_n','b_n','p_n','f(p_n)'])
display(df)
print(f'RESULT p_{i} = {p}')
```
