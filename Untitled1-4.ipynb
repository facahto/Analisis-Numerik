```python
def f_real(x,t):
    d1_ = (torch.log(x/K) + (rbs + 0.5*(sigma**2))*(T_-t))/(sigma*torch.sqrt(T_-t))
    d2_ = d1_ - (sigma*torch.sqrt(T_-t))
    Nd1 = norm.cdf(d1_)
    Nd2 = norm.cdf(d2_)
    return x*Nd1 - (K*torch.exp(-rbs*(T_-t))*Nd2)
```

```python
def forward(X, p_unit):
    
    activate = nn.Tanh()

    ni = layers[0]
    n1 = layers[1]
    no = layers[-1]

    a1 = n1*ni
    a2 = (no*n1) + a1

    d1 = (dim-b)+(n1*1)
    d2 = (no*1) + d1


    w0 = torch.reshape(p_unit[:a1],(n1,ni))
    w1 = torch.reshape(p_unit[a1:a2],(no,n1))

    b0 = torch.reshape(p_unit[dim-b:d1],(n1,1))
    b1 = torch.reshape(p_unit[d1:d2],(no,1))


    Z1 = X@w0.T + b0.T
    A1 = activate(Z1)
    Z2 = A1@w1.T + b1.T
    
    return Z2
```
